```python
def gender_features(word):
    return {'last_letter': word[-1]}

gender_features('Shrek')
```

```python
male_names_cls = [(name, 'male') for name in names.words('male.txt')]
female_names_cls = [(name, 'female') for name in names.words('female.txt')]

# Combine the lists.
labeled_names = male_names_cls + female_names_cls

# Shuffle the list.
random.shuffle(labeled_names)
```

```python
from nltk import NaiveBayesClassifier

# Extract the features using the `gender_features()` function.
featuresets = [(gender_features(n), gender) for (n, gender) in labeled_names]

# Split the dataset into train and test set.
train_set, test_set = featuresets[500:], featuresets[:500]
print(train_set)
```

```python
from gensim import utils
from gensim.models import Word2Vec
import sys
import os
class MyCorpus:
  def __iter__(self):
    for word in names.words():     
      yield utils.simple_preprocess(word)
      
```

```python
import gensim 
Sentences = MyCorpus()
print(Sentences)
model = gensim.models.Word2Vec(
        sentences=Sentences,
        size=150,
        window=10,
        min_count=1)
print(model)
```

```python
vocab_words = []
vocab_words = model.wv.vocab
print(vocab_words)
```

```python
from sklearn import cluster
from sklearn import metrics
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

X = model[vocab_words]

X= np.array(X)
visualization_fit = cluster.KMeans(n_clusters=10, random_state=0).fit_transform(X)
```

```python
from numpy import array
from sklearn.feature_extraction.text import CountVectorizer
from keras.utils import np_utils
import tensorflow as tf
import numpy as np
import subprocess, re
import tensorflow.python.framework.ops
from tensorflow.python.framework.ops import EagerTensor


cv = CountVectorizer(binary=True,lowercase=False, min_df=0, max_df=1, ngram_range=(1,1))

cv_train_reviews = cv.fit_transform(X_train)

cv_test_reviews = cv.transform(X_test)

cv_train_reviews.shape
```

```python
print(len(y_test))
print(len(y_train))
```

```python
from sklearn.preprocessing import LabelBinarizer
lb=LabelBinarizer()

train_names_data=lb.fit_transform(y_train)

test_names_data=lb.transform(y_test)
print(train_names_data)
print(test_names_data)
```

```python
cv_train_reviews.shape
```

```python
cv_test_reviews.shape
```

```python
cv_train_reviews = cv_train_reviews.reshape(cv_train_reviews.shape[0], 3722)

cv_test_reviews = cv_test_reviews.reshape(cv_test_reviews.shape[0], 3722)

cv_train_reviews.shape
cv_test_reviews.shape
```

```python
cv_train_reviews= cv_train_reviews.astype('float32')
cv_test_reviews= cv_test_reviews.astype('float32')

cv_train_reviews/=255.0
cv_test_reviews/=255.0
print(cv_train_reviews.shape)
print(cv_test_reviews.shape)
```

```python
train_names_data.shape
```

```python
train_names_data = np.vstack(train_names_data)
```

```python
test_names_data.shape
```

```python
test_names_data = np.vstack(test_names_data)
```

```python
from keras.utils import np_utils
import tensorflow as tf
no_of_classes = 10
y_train_data= np_utils.to_categorical(train_names_data, no_of_classes)
y_test_data= np_utils.to_categorical(test_names_data,no_of_classes)
y_train_data= tf.stack(y_train_data)
y_test_data = tf.stack(y_test_data)
print(y_train_data.shape)
print(y_test_data.shape)
```

```python
from keras.engine.sequential import Sequential
from keras.layers import Dense, Dropout, Flatten, Activation
from keras.layers import Conv1D, MaxPooling1D
import keras
import tensorflow as tf
from tensorflow.keras.layers import BatchNormalization

model = Sequential()
model.add(tf.keras.layers.InputLayer(input_shape=(3722, 1)))
model.add(Conv1D(64, 8, activation='sigmoid'))
model.add(BatchNormalization(axis=-1))
model.add(Conv1D(64, 4, activation='sigmoid'))
#model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))
model.add(MaxPooling1D(pool_size=2))



model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['accuracy'])
```

```python
model.fit(cv_train_reviews, cv_test_reviews, batch_size=128, epochs=5, verbose=1, validation_data=(y_train_data, y_test_data))
```
