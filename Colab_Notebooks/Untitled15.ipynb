```python
from urllib import request 
from bs4 import BeautifulSoup
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
def content_of_URL(URL):
    tags = request.urlopen(URL).read().decode('utf8')
    raw_data = BeautifulSoup(tags).get_text()
    print(raw_data)
    tok = word_tokenize(raw_data)
    return tok
```

```python
content_of_URL(
    'http://www.nltk.org/')
```
