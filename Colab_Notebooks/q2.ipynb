```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
 # Download NLTK models
nltk.download('stopwords')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
nltk.download('maxent_ne_chunker')
```

```python
def nameEntityRecog(txt):
   
  
    # Tokenize the text
    word_tokens = word_tokenize(txt)
  
   
    stopWords = set(stopwords.words('english'))
    token_words = [token for token in word_tokens if token.lower() not in stopWords]

    # Performing PartsofSpeech tag
    tagged_words = nltk.pos_tag(token_words)

    # Performing NamedEntityRecognition
    chunk = nltk.ne_chunk(tagged_words, binary=False)

    # Extract named entities and their types
    namedEntities = []
    for tree in chunk:
        if hasattr(tree, 'label') and tree.label() == 'NE':
            entityName = ' '.join(child[0] for child in tree.leaves())
            entityType = tree.label()
            namedEntities.append((entityName, entityType))

    return namedEntities
```

```python
nltk.download("brown")
from nltk.corpus import brown
corpus_text = " ".join(brown.words()[:1000000])  # You can adjust the number of words as needed
result = nameEntityRecog(corpus_text)
print(result)
```
