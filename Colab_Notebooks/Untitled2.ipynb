```python
import time
def insertionSort(arr):
  
    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
  
        key = arr[i]
  
        # Move elements of arr[0..i-1], that are
        # greater than key, to one position ahead
        # of their current position
        j = i-1
        while j >=0 and key < arr[j] :
                arr[j+1] = arr[j]
                j -= 1
        arr[j+1] = key
    
        
import numpy as np

n= [80000,100000]
l1=[]
for j in range(len(n)):
    arr = np.arange(n[j],0,-1)
    exe_time = 0
    for i in range(5):
        start = time.time()
        insertionSort(arr)
        end = time.time()
        exe_time = (end - start) + exe_time
    avg_exe_time = exe_time/5
    l1.append(avg_exe_time)
    print(avg_exe_time)
print(l1)

```
