```python
import nltk
from collections import Counter
from nltk.corpus import stopwords
from nltk import word_tokenize, pos_tag, RegexpParser

```

```python
nltk.download('punkt')

nltk.download('stopwords')
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
def topTen_nounPhrases(txt, N=10):
    #Tokenizing the text
    words = word_tokenize(txt)

    
    stopWords = set(stopwords.words('english'))
    filteredWords = [wrd for wrd in words if wrd.lower() not in stopWords]

   
    taggedWords = pos_tag(filteredWords)

    
    gram = "NP: {<DT>?<JJ>*<NN.*>+}"

    
    chunkParse = RegexpParser(gram)
    tree = chunkParse.parse(taggedWords)

    
    nounPhrase = []
    for subtre in tree.subtrees(filter=lambda tr: tr.label() == 'NP'):
        np = " ".join([word for word, tag in subtre.leaves()])
        nounPhrase.append(np)

    
    return Counter(nounPhrase).most_common(N)
```

```python
from nltk.corpus import brown
nltk.download('brown')


txt = " ".join(brown.words())


output = topTen_nounPhrases(txt)


```

```python
from tabulate import tabulate

table = [[x, y] for x, y in output]
column_name = ['np', 'Count']
print("The top 10 Noun_Phrases in BrownCorpus:")

print(tabulate(table, headers=column_name))
```
