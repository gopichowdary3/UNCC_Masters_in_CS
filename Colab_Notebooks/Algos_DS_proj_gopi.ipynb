```python
import time
def insertionSort(arr):
  
    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
  
        key = arr[i]
  
        # Move elements of arr[0..i-1], that are
        # greater than key, to one position ahead
        # of their current position
        j = i-1
        while j >=0 and key < arr[j] :
                arr[j+1] = arr[j]
                j -= 1
        arr[j+1] = key
    
        
import numpy as np

n= [1000,2000,4000,5000,10000,20000,40000,50000,60000,80000,100000]
l1=[]
for j in range(len(n)):
    arr = np.random.randint(1,500, n[j])
    exe_time = 0
    for i in range(5): #this for loop finds 5 execution times of the sorting
        start = time.time()
        insertionSort(arr)
        end = time.time()
        exe_time = (end - start) + exe_time #execution time of sorting
    avg_exe_time = exe_time/5 #average execution time of sorting
    l1.append(avg_exe_time)

```

```python
# Python program for implementation of MergeSort
def merge(arr, l, m, r):
    n1 = m - l + 1
    n2 = r - m
 
    # create temp arrays
    L = [0] * (n1)
    R = [0] * (n2)
 
    # Copy data to temp arrays L[] and R[]
    for i in range(0, n1):
        L[i] = arr[l + i]
 
    for j in range(0, n2):
        R[j] = arr[m + 1 + j]
 
    # Merge the temp arrays back into arr[l..r]
    i = 0     # Initial index of first subarray
    j = 0     # Initial index of second subarray
    k = l     # Initial index of merged subarray
 
    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1
 
    # Copy the remaining elements of L[], if there
    # are any
    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1
 
    # Copy the remaining elements of R[], if there
    # are any
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1
 
# l is for left index and r is right index of the
# sub-array of arr to be sorted
 
 
def mergeSort(arr, l, r):
    if l < r:
 
        # Same as (l+r)//2, but avoids overflow for
        # large l and h
        m = l+(r-l)//2
 
        # Sort first and second halves
        mergeSort(arr, l, m)
        mergeSort(arr, m+1, r)
        merge(arr, l, m, r)

l2 = []  
for j in range(len(n)):
    arr = np.random.randint(0,500,n[j])
    exe_time = 0
    for i in range(5):
        start = time.time()
        mergeSort(arr,0,len(arr)-1)
        end = time.time()
        exe_time = (end - start) + exe_time
    avg_exe_time= exe_time/5
    print(avg_exe_time)
    l2.append(avg_exe_time)
print(l2)
```

```python
#HeapSort
def heapify(arr, n, i):
    largest = i  # Initialize highest as root
    l = 2 * i + 1     # for a given node, left child is at position = 2*i + 1
    r = 2 * i + 2     # for a given node, right child is at position = 2*i + 2
 
    # find if left child of root exists and is greater than root
    if l < n and arr[largest] < arr[l]:
        largest = l
 
    # find if right child of root exists and is greater than root
    if r < n and arr[largest] < arr[r]:
        largest = r
 
    # Change root, if necessesary
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # swapping
        heapify(arr, n, largest)# Heapifying the root.
 
def heapSort(arr):
    n = len(arr)
 
    # Build a maxheap.
    for i in range(n//2 - 1, -1, -1):
        heapify(arr, n, i)
 
    # One by one extract elements
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  # swap
        heapify(arr, i, 0)

l3 = []  
for j in range(len(n)):
    arr = np.random.randint(1,500, n[j])
    exe_time = 0
    for i in range(5):
        start = time.time()
        heapSort(arr)
        end = time.time()
        exe_time = (end - start) + exe_time
    avg_exe_time= exe_time/5
    print(avg_exe_time)
    l3.append(avg_exe_time)
```

```python
import random
import time

def inplaceQuickSort(arr, l=0, r=None):
    # if r is None then take the last index of an array
    if r is None:
        r = len(arr) - 1
    # if l crosses r then return
    if l >= r:
        return
    # take random int from l to r and use that as pivot
    pivotInd = random.randint(l,r)
    # call partition function for pivotInd and take the position of the pivot after successful partition
    i = inplacePartition(arr, l, r, pivotInd)

    inplaceQuickSort(arr, l, i-1)
    inplaceQuickSort(arr, i+1, r)
    return arr

def inplacePartition(arr, l, r, pivot):
    temp = l
    # swapping first element with pivot
    arr[l], arr[pivot] = arr[pivot], arr[l]
    pivotElem = arr[l]
    # iterating through all elements and swapping then with temp position when we find element < pivot
    for i in range(l+1, r+1):
        if arr[i] <= arr[l]:
            temp += 1   # just to keep track of placed elements
            arr[i], arr[temp] = arr[temp], arr[i]
    # swap back the pivot to its actual position where left is less valued and right is more values elements.
    arr[temp], arr[l] = arr[l], arr[temp]
    return temp

l4 = [] 
n= [1000,2000,4000,5000,10000,20000,40000,50000,60000,80000,100000]
for j in range(len(n)):
    arr = np.random.randint(1,500, n[j])
    exe_time = 0
    for i in range(5):
        start = time.time()
        inplaceQuickSort(arr)
        end = time.time()
        exe_time = (end - start) + exe_time
    avg_exe_time= exe_time/5
    print(avg_exe_time)
    l4.append(avg_exe_time)
print(l4)
  
```
