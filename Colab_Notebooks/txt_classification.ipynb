```python
#1.Selecting any four categories from the dataset
categories = ['alt.atheism','comp.windows.x','rec.autos','sci.crypt']

#1.Downloading the train and test data for those four categories
train_data = fetch_20newsgroups(subset='train',remove=('headers', 'footers', 'quotes'), categories=categories) 
test_data = fetch_20newsgroups(subset='test',remove=('headers', 'footers', 'quotes'),categories=categories)

```
