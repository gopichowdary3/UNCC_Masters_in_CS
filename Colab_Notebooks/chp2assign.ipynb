```python
nltk.download('state_union')
from nltk.corpus import state_union
```

```python
state_union.fileids()
```

```python
cond_freq_dist = nltk.ConditionalFreqDist((target,fileid[:4])for fileid in state_union.fileids() for w in state_union.words(fileid) for target in  ['men','women', 'people'] if w.lower().startswith(target))
```

```python
cond_freq_dist['people']
```

```python
cond_freq_dist['men']
```

```python
cond_freq_dist['women']
```

```python
cond_freq_dist.plot()
```

 What has happened to the usage of these words over time
 the usage of people has increase in the years btwn : 1990 nad 1995
 at the beginning the usage of word 'women' is less compared to 'men'
 however at the later ages both words usage is almost similar
