```python
import keras
import tensorflow as tf
data = tf.keras.datasets.mnist.load_data()
```

```python
(X_train, y_train), (X_test, y_test) = data
print(X_train)
print(X_test)
```

```python
X_train.shape
num_pixels = 784
X_train = X_train.reshape(X_train.shape[0],
                         num_pixels)
X_test = X_test.reshape(X_test.shape[0],
                         num_pixels)
print(X_train.shape)
```
