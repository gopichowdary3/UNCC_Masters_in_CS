```python
import numpy as np
word_embeddings = dict()
x=[]

with open('/content/drive/MyDrive/NLP/5111_4111Spring2023Shared-20230125T210009Z-001/5111_4111Spring2023Shared/Copy of glove.6B.50d.txt', encoding='utf-8') as f:
    for line in f:
        word = line.split()[0]
        x.append(word)
        embeddings = np.asarray(line.split()[1:], dtype='float32')
        word_embeddings[word] = embeddings


wrd = word_embeddings
```
