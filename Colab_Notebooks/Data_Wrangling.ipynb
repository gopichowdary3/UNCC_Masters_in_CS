```python
import urllib
import urllib.request
import pandas as pd
dataset = urllib.request.urlopen('https://gist.githubusercontent.com/jvilledieu/c3afe5bc21da28880a30/raw/a344034b82a11433ba6f149afa47e57567d4a18f/Companies.csv')
df = pd.read_csv(dataset)
df
```

```python
df2 = pd.read_html('https://en.wikipedia.org/wiki/List_of_chief_executive_officers#List_of_CEOs')[0]
df2
```

```python
crawl_list=[]

for executive in df2['Executive']:
    crawl_list.append('https://en.wikipedia.org/wiki/'+executive.replace(' ','_'))
crawl_list
```

```python
#the URL link from which the DOB was retrieved.
df2['WikiPage'] = crawl_list
df2
```

```python
import requests
from bs4 import BeautifulSoup

crawl_list1=[]

for pages in df2['WikiPage']:
    url=pages
    source=requests.get(url)
    plain=source.text
    soup=BeautifulSoup(plain)
    birth_date=soup.find("span",{"class":"bday"})
    birth_date = 'NA' if birth_date is None else birth_date
    crawl_list1.append(birth_date)
```

```python
df2['DOB'] = crawl_list1
df2
```
